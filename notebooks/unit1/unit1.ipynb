## 深度强化学习简要回顾 📚

<img src="https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/unit1/RL_process_game.jpg" alt="The RL process" width="100%">

让我们回顾一下我们在第一单元中学到的知识：

- **强化学习**是一种**从动作中学习的计算方法**。我们构建一个智能体，通过**试错**与环境交互并接收奖励（负奖励或正奖励）作为反馈来学习。

- 任何 RL 智能体的目标都是**最大化其预期累积奖励**（也称为预期回报），因为 RL 基于*奖励假设*，即所有目标都可以描述为最大化预期累积奖励。

- RL 过程是一个**循环，输出状态、动作、奖励和下一个状态的序列**。

- 为了计算预期累积奖励，**我们会对奖励进行折扣**：越早出现的奖励（游戏开始时）越容易发生，因为它们比长期未来的奖励更可预测。

- 要解决 RL 问题，你需要**找到一个最优策略**；策略是 AI 的“大脑”，它告诉我们在给定状态下该采取什么动作。

寻找最优策略有两种主要方式：
- **直接训练策略**：基于策略的方法 (Policy-based methods)。
- **训练价值函数**：告诉智能体在每个状态下的预期回报，并使用此函数来定义策略：基于价值的方法 (Value-based methods)。

- 最后，我们提到了**深度 RL**，因为我们引入了**深度神经网络来估计要采取的动作（基于策略）或估计状态的价值（基于价值）**。

## 了解 Gymnasium 及其运作方式 🤖

🏋 包含我们环境的库叫做 Gymnasium。
**在深度强化学习中你会经常使用 Gymnasium。**

Gymnasium 是由 Farama 基金会维护的 Gym 库的**新版本**。

Gymnasium 库提供两样东西：
- 一个允许你**创建 RL 环境**的接口。
- 一个**环境集合**（如 Box2D、Atari 等）。

在每一步中：
- 智能体从**环境**接收**状态 (S0)**——我们收到游戏的第一帧。
- 基于该**状态 (S0)**，智能体采取一个**动作 (A0)**——我们的智能体向右移动。
- 环境转换到**新状态 (S1)**——新帧。
- 环境给智能体一些**奖励 (R1)**——我们没死（*正奖励 +1*）。

使用 Gymnasium：
1️⃣ 使用 `gymnasium.make()` 创建环境
2️⃣ 使用 `observation, info = env.reset()` 将环境重置为初始状态
3️⃣ 获取动作（例如使用模型预测，或者像下例中采取随机动作）
4️⃣ 使用 `env.step(action)` 执行动作并获得结果。

````python
import gymnasium as gym

# First, we create our environment called LunarLander-v3
env = gym.make("LunarLander-v3")

# Then we reset this environment
observation, info = env.reset()

for _ in range(20):
  # Take a random action
  action = env.action_space.sample()
  print("Action taken:", action)

  # Do this action in the environment and get
  # next_state, reward, terminated, truncated and info
  observation, reward, terminated, truncated, info = env.step(action)

  # If the game is terminated (in our case we land, crashed) or truncated (timeout)
  if terminated or truncated:
      # Reset the environment
      print("Environment is reset")
      observation, info = env.reset()

env.close()
````

## 创建月球着陆器环境 🌛 并了解其运作方式

在这个教程中，我们将训练一个[月球着陆器](https://gymnasium.farama.org/environments/box2d/lunar_lander/)**正确降落在月球上**。为此，智能体需要学习**调整速度和位置（水平、垂直和角度）来正确着陆。**

让我们看看环境是什么样的：

````python
# We create our environment with gym.make("<name_of_the_environment>")
env = gym.make("LunarLander-v3")
env.reset()
print("_____OBSERVATION SPACE_____ \n")
print("Observation Space Shape", env.observation_space.shape)
print("Sample observation", env.observation_space.sample()) # Get a random observation
````

通过 `Observation Space Shape (8,)` 我们看到观测是一个长度为 8 的向量，每个值包含关于着陆器的不同信息：
- 水平坐标 (x)
- 垂直坐标 (y)
- 水平速度 (x)
- 垂直速度 (y)
- 角度
- 角速度
- 左腿是否接触地面（布尔值）
- 右腿是否接触地面（布尔值）

````python
print("\n _____ACTION SPACE_____ \n")
print("Action Space Shape", env.action_space.n)
print("Action Space Sample", env.action_space.sample()) # Take a random action
````

动作空间（智能体可以采取的动作集合）是离散的，包含 4 个动作 🎮：

- 动作 0：无操作
- 动作 1：发射左侧引擎
- 动作 2：发射主引擎
- 动作 3：发射右侧引擎

**奖励函数**（每一步给予反馈的函数）💰：

每一步后都会获得奖励。一个回合的总奖励是该回合内**所有步数奖励的总和**。

具体奖励规则：
- 着陆器离着陆垫越近/越远，奖励增加/减少。
- 着陆器移动越慢/越快，奖励增加/减少。
- 着陆器倾斜角度越大（非水平），奖励减少。
- 每条腿接触地面奖励 10 分。
- 侧边引擎发射时每帧扣除 0.03 分。
- 主引擎发射时每帧扣除 0.3 分。

**坠毁或安全着陆将额外获得 -100 或 +100 分。**

如果一个回合得分**至少达到 200 分，则被视为解决了该环境。**

````python
from huggingface_sb3 import load_from_hub
repo_id = "Classroom-workshop/assignment2-omar" # The repo_id
filename = "ppo-LunarLander-v2.zip" # The model filename.zip

# When the model was trained on Python 3.8 the pickle protocol is 5
# But Python 3.6, 3.7 use protocol 4
# In order to get compatibility we need to:
# 1. Install pickle5 (we done it at the beginning of the colab)
# 2. Create a custom empty object we pass as parameter to PPO.load()
custom_objects = {
            "learning_rate": 0.0,
            "lr_schedule": lambda _: 0.0,
            "clip_range": lambda _: 0.0,
}

checkpoint = load_from_hub(repo_id, filename)
model = PPO.load(checkpoint, custom_objects=custom_objects, print_system_info=True)
````
